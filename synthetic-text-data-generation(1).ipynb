```python

# ✅ Step 2: Import libraries
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from transformers import BitsAndBytesConfig

# ✅ Step 3: Define the model name (You can change this to another LLM)
model_name = "meta-llama/Llama-3.2-3B-Instruct"  # Change if needed

# ✅ Step 4: Configure 4-bit quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,                # Enable 4-bit quantization
    bnb_4bit_compute_dtype=torch.float16,  # Use FP16 for efficiency
    bnb_4bit_use_double_quant=True,   # Further reduces memory usage
)

# ✅ Step 5: Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)

# ✅ Step 6: Load model with 4-bit quantization
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,  # Apply BitsAndBytes config
    device_map="auto",               # Automatically assigns GPU/CPU
    trust_remote_code=True
)

# ✅ Step 7: Create text-generation pipeline
llm = pipeline("text-generation", model=model, tokenizer=tokenizer)

# ✅ Step 8: Define the prompt
```

```python
prompt = """
Generate a corporate training script for a senior data visualisation engineer
Include a trainer and a trainee discussing how to build advanced and interactive data visualization using Tableau/PowerBI.
Provide a realistic conversation format.
"""

# ✅ Step 9: Generate text
generated_text = llm(
    prompt, 
    max_length=800, 
    num_return_sequences=1, 
    do_sample=True
)[0]["generated_text"]

# ✅ Step 10: Print the generated script
print("\nGenerated Training Script:\n")
print(generated_text)

# ✅ Step 11: Save output to a text file
with open("training_script.txt", "w") as f:
    f.write(generated_text)

print("\n✅ Synthetic training script saved as 'training_script.txt'")
```
